## Comparison between Profit and Quantity

```python
#Plotting the chart for Profit vs Quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df1)
plt.legend()
plt.show()
```

```python
#Plotting the sales and profit for each segment
df1.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
#Plotting the sales of different segments according to the regions
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df1,hue='Segment',order=df1['Region'].value_counts().index,palette='Spectral')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
#Plotting Profit and Sales distribution, region-wise 
df1.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','green'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
t_states = df1['State'].value_counts().nlargest(10)
t_states
```

```python
df1.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],alpha=0.9,figsize=(10,8))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Technology and Office Supplies have higher profits whereas Furniture has less profits.

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','green'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

#### Conclusion:
1.Phones category has the highest sales
2.Chairs have high sales but comparitively less profit than phones.
3.Tables and Bookcases category are facing losses.
4.Envelopes, Labels and Fastners contribute the least towards sales and profit.
5.Paper, Copiers and Envelopes have higher profit margin.
